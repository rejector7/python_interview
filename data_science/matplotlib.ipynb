```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 2*np.pi, 20)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure(figsize=(14, 10))
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, y1)
ax1.plot(x, y2)
ax1.grid(True, color='green', linestyle='-.')
# ax1.title('pic1', color='yellow', loc='right')
plt.title('pic1', color='red', loc='right')


ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, y1)
ax2.plot(x, y2)
ax2.grid(False)
plt.axis('off')
plt.title('pic2', color='red', loc='center')

ax3 = fig.add_subplot(2, 2, 3, facecolor='#FFDAB9')
ax3.plot(x, y1)
ax3.plot(x, y2)
plt.xlim([0, 6])
plt.ylim([-1.5, 1.5])
plt.xlabel('X', size=15, color='red')
plt.ylabel('Y', size=15, color='red', rotation=0)
plt.title('pic3', color='red', loc='center')

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(x, y1, label='sin(x)', ls='--')
ax4.plot(x, y2, label='cos(x)', marker='o')
plt.axis('equal')
plt.legend(loc=2)
plt.title('pic4', color='red', loc='center')
plt.annotate(
    s="point high",
    xytext=(3, 1.5),
    xy=(1.5, 1.2),
    arrowprops={
        'width':5,
        'headlength':20,
        'headwidth':10
    }
)

# plt.savefig('pic.jpg')




```
